# Week 3 - Tutorial 3 Documentation

## 1. Identify the Components
* **Inputs:** 
  * Quiz 1 mark (`quiz_1`)
  * Quiz 2 mark (`quiz_2`)
  * Quiz 3 mark (`quiz_3`)
  * User decision to continue (`choice`)
* **Process:**
  * Calculate the average: $Average = \frac{Quiz1 + Quiz2 + Quiz3}{3}$
  * Check the condition if $Average \ge 50$ to determine pass or fail status.
* **Output:**
  * The calculated average mark.
  * The status text ("Pass" or "Fail").
  * Termination message ("Program Ended").

## 2. Design the Algorithm
### 2.1 Pseudocode
```text
BEGIN
    SET choice = "Y"
    WHILE choice == "Y" OR choice == "y" DO
        INPUT quiz_1
        INPUT quiz_2
        INPUT quiz_3
        
        COMPUTE average = (quiz_1 + quiz_2 + quiz_3) / 3
        OUTPUT average
        
        IF average >= 50 THEN
            OUTPUT "Pass"
        ELSE
            OUTPUT "Fail"
        ENDIF
        
        INPUT choice
    ENDWHILE
    OUTPUT "Program Ended"
END

```

### 2.2 Algorithm Structure
* **Sequence:** Used inside the loop to get inputs, compute the average, and display the result step-by-step.
* **Selection:** Used in the `if-else` statement to decide whether the student passed or failed based on the score.
* **Iteration:** Used in the `while` loop to repeat the process for multiple students until the user decides to stop.
